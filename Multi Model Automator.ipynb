```python
import pandas as pd
import numpy as np
df=pd.read_csv("cardio_train.csv")
df
```

```python
df_p=pd.DataFrame(df)
df_p.describe()
```

```python
#Systolic blood pressure
for i in range (len(df['ap_hi'])):
    if df['ap_hi'][i]<0:
        df['ap_hi'][i]=df['ap_hi'].mean()
    elif df['ap_hi'][i]> 220:
        df['ap_hi'][i]=df['ap_hi'].std()
    else:
        pass
    
```

```python
# Diastolic Blood Pressure
for i in range (len(df['ap_lo'])):
    if df['ap_lo'][i]<0:
        df['ap_lo'][i]=df['ap_lo'].mean()
    elif df['ap_lo'][i]> 140:
        df['ap_lo'][i]=df['ap_lo'].std()
    else:
        pass
    
```

```python
# Age To Years
for i in range(len(df['age'])):
    df['age'][i]=df['age'][i]/365
    
```

```python
y=df['cardio']
x=df.drop(['cardio'],axis=1)
```

```python
x.hist()
```

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts=train_test_split(x,y,test_size=0.25,random_state=23)


    # In[9]:


   


    # In[22]:
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtr,ytr)


    # In[23]:



```

```python
yp=lr.predict(xts)


    # In[24]:

yp_acc2=lr.score(xts,yts)
print("accuracy :",yp_acc2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
#X, y = make_classification(n_samples=1000, n_features=4,n_informative=2, n_redundant=0,random_state=0, shuffle=False)
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(xtr, ytr)
yp=clf.predict(xts)
```

```python
from xgboost import XGBClassifier
model=XGBClassifier()
model.fit(xtr,ytr)
#Evaluate the dec tree
print("Accuracy:",model.score(xts,yts))
yp=model.predict(xts)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model=GradientBoostingClassifier(n_estimators=100,max_depth=5)
model.fit(xtr,ytr)
#Evaluate the dec tree
print("Accuracy:",model.score(xts,yts))
yp=model.predict(xts)
```
